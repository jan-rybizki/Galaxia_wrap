```python
ext = np.load('ext_sloan.npy')
ext_std = np.load('ext_std_sloan.npy')
parsec = np.load('parsec_sloan.npy')
parsec_std = np.load('parsec_std_sloan.npy')
```

```python
for t in [5,50,95,99,100]:
    print(t)
    print(np.percentile(ext_std,t,axis = 0))
```

```python
for item in parsec.dtype.names:
    print(item)
    for t in [5,50,95,99,100]:
        print(t,np.percentile(parsec_std[item],t))
    print("#########################")
```

```python
# We use a bigger font for you to see better on the beamer
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 22}
matplotlib.rc('font', **font)

## We have a problem with those high extincted stars. They have the same properties as the rest
## but strongly diverging photometry. Could be a problem when mapping to other photometry. Write into paper.

sort = np.argsort(parsec_std["sloan_g"])
plt.figure(figsize=(20,10))
plt.scatter(parsec["log_teff"][sort],parsec["log_lum"][sort],c = parsec_std['sloan_g'][sort])
plt.colorbar(label="std in sloan_g")
plt.gca().invert_xaxis()
plt.ylabel("loglum")
plt.xlabel("logteff")
```
